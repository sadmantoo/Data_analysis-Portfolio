# HIV/AIDS Analysis
### HIV/AIDS remains a significant public health challenge globally, with sub-Saharan Africa bearing the highest burden of the disease. Among the countries grappling with this epidemic, South Africa stands out, facing complex socio-economic and healthcare challenges in combating HIV/AIDS. This analysis delves into various dimensions of the HIV/AIDS epidemic in South Africa, leveraging a diverse array of datasets sourced from reputable sources. By examining epidemiological trends, prevention strategies, treatment access, mortality rates, resource allocation, and public awareness, this study aims to provide a comprehensive understanding of the current landscape of HIV/AIDS in South Africa. Through data-driven insights, policymakers, healthcare professionals, and stakeholders can better tailor interventions, allocate resources effectively, and advance towards the goal of controlling and ultimately ending the HIV/AIDS epidemic in South Africa.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
df = pd.read_csv('new-cases-of-hiv-infection.csv')
df.shape
df.head(10)
```

# Epidemiological Analysis:

1. What is the trend of new HIV infections in South Africa over the past decade?
2. What is the age distribution of HIV-related deaths in South Africa?

```python
df_zar = df [df['Entity'] == 'South Africa']
df_zar.head(10)
```

```python
df_zar_last_10_years = df_zar[df_zar['Year'] >= df_zar['Year'].max() - 9]  # Filter last 10 years
ax = df_zar_last_10_years.set_index('Year').plot(figsize=(16,10))
ax.set_xlabel("Year")
ax.set_ylabel("New HIV Infections")
ax.set_title("Trend of New HIV Infections in South Africa Over the Past Decade")
plt.show()
```

### ANALYSIS

Over the last decade we can see that the number of new HIV infections has been on a decline. This may be due to a number of factors such as increased awareness and education about HIV/AIDS prevention, improved access to testing and treatment services, implementation of effective prevention strategies including condom use promotion and voluntary medical male circumcision, as well as targeted interventions for key populations at higher risk of HIV transmission such as sex workers and men who have sex with men.

2. What is the age distribution of HIV-related deaths in South Africa?

```python
df2 = pd.read_csv('deaths-from-hiv-by-age.csv')
df2_zar = df2[df2['Entity'] == 'South Africa']
df2_zar.head(10)
```

```python
ax = df2_zar.set_index('Year').plot(figsize=(16,10))
ax.set_xlabel("Year")
ax.set_ylabel("Number of Deahead")
ax.set_title("Age distribution of HIV-related deaths in South Africa")
plt.show()
```

### ANALYSIS

So starting from 1990s, the number of HIV-ralted deaths was highest in 2005 and the age group (15-49) were the most effected. The age groups that are least affected are kids ranging from 5-14 and adults in the 70+ range.

```python
df5 = pd.read_csv('tb-patients-tested-positive-for-hiv.csv')
df5_zar = df5[df5['Entity'] == 'South Africa']
df5_zar.head(10)
```

```python
df6 = pd.read_csv('tb-patients-living-with-hiv-receiving-art.csv')
df6_zar = df6[df6['Entity'] == 'South Africa']
df6_zar.head(10)
```

```python
df5_zar_filtered = df5_zar[df5_zar['Year'] >= 2003]
df6_zar_filtered = df6_zar[df6_zar['Year'] >= 2003]

merged_df = pd.merge(df5_zar_filtered, df6_zar_filtered, on=["Entity", "Code", "Year"], suffixes=('_df5', '_df6'))
merged_df.head(10)
```

```python
merged_df['Proportion Receiving ART'] = (merged_df['TB patients living with HIV receiving ART - Total'] / merged_df['TB patients tested positive for HIV - Total']) * 100
average_art_percentage = merged_df['Proportion Receiving ART'].mean()
print("Average proportion of TB patients receiving ART among those tested positive for HIV in South Africa:", average_art_percentage, "%")
```

```python
plt.figure(figsize=(10, 6))
plt.bar(merged_df['Year'], merged_df['Proportion Receiving ART'], color='lightgreen')
plt.axhline(y=average_art_percentage, color='red', linestyle='--', label=f'Average: {average_art_percentage:.2f}%')
plt.title('Proportion of TB Patients Receiving ART Among Those Tested Positive for HIV')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
df8 = pd.read_csv('number-of-children-orphaned-from-aids.csv')
df8_zar = df8[df8['Entity'] == 'South Africa']
df8_zar.head(10)
```

```python
total_orphans_per_year = df8_zar.groupby('Year')['AIDS Orphans - 0-17 estimate'].sum()

plt.figure(figsize=(10, 6))
plt.plot(total_orphans_per_year.index, total_orphans_per_year.values, marker='o', linestyle='-')
plt.title('Trend of Children Orphaned due to AIDS in South Africa')
plt.xlabel('Year')
plt.ylabel('Number of Children Orphaned')
plt.grid(True)
plt.xticks(rotation=45)

formatter = FuncFormatter(lambda x, _: '{:,.0f}'.format(x))
plt.gca().yaxis.set_major_formatter(formatter)

plt.tight_layout()
plt.show()
```
